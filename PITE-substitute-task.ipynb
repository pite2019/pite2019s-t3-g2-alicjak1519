```python
import sklearn
import sklearn.datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
bostonDataset = sklearn.datasets.load_boston();
```

```python
bostonAsPanda = pd.DataFrame(bostonDataset.data, columns = [bostonDataset.feature_names]);
```

```python
bostonAsPanda.head()
```

```python
bostonAsPandaX = bostonAsPanda.loc[:,["CRIM","AGE"]]
```

```python
bostonAsPandaX.head()
```

```python
bostonAsPandaX.hist(bins = 10);
```

```python
from sklearn.linear_model import LinearRegression

X1 = bostonAsPandaX["AGE"]
X2 = bostonAsPandaX["CRIM"]
y = bostonDataset.target

reg1 = LinearRegression().fit(X1,y)
reg2 = LinearRegression().fit(X2,y)

print(reg1.score(X1,y))
print(reg2.score(X2,y))

predicted1 = sklearn.model_selection.cross_val_predict(reg1, bostonAsPandaX, y, cv = 5);
predicted2 = sklearn.model_selection.cross_val_predict(reg2, bostonAsPandaX, y, cv = 5);

# fig, ax = plt.subplots()
# ax.plot(X1, y, 'g--')
# ax.set_xlabel('Measured Age')
# ax.set_ylabel('Predicted Age')
# plt.show()

fig, ax = plt.subplots()
ax.scatter(y, predicted1, edgecolors=(0,1,0))
ax.set_xlabel('Measured Age')
ax.set_ylabel('Target')
plt.show()


# fig, bx = plt.subplots()
# bx.plot(X2,y,'r--')
# bx.set_xlabel('Measured Crime')
# bx.set_ylabel('Predicted Crime')
# plt.show()

fig, bx = plt.subplots()
bx.scatter(y, predicted2, edgecolors=(1,0,0))
bx.set_xlabel('Measured Crime')
bx.set_ylabel('Target')
plt.show()
```
